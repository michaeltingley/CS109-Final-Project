# Purpose
The purpose of this project is to develop machine learning tools for chemical data. Of particular interest are energetic properties that may help predict a molecule's ability to act as an organic photocell. Experimental validation of a candidate organic photovoltaic molecule is an expensive and lengthy process, so the ability to screen promising candidates is very desirable. Currently, quantum calculations take hours to compute properties for a single molecule. Our goal is to learn models that will quickly approximate such calculations, in order to improve the throughput of candidate molecule proposal.

# Motivation
This project is motivated largely by the efforts of the Harvard Clean Energy Project. This is a research group on campus at the intersection of chemistry, biology, and computation, whose goals are to improve the theoretical understanding and practical manipulation of organic materials in order to provide, among other things, sustainable clean energy through efficient use of materials science in the area of photovoltaics.

One important area of particular interest presently is improving the materials that are used in the construction of solar cells. Currently, solar cells are predominantly manufactured using silicon, for the principal reason that they have high relative conversion efficiencies of around 15% (Riede et al., 2008). They also can have long practical lifetimes of around 25 years. However, the downsides to using silicon are numerous: (1) the material is expensive, (2) it must be manufactured in a controlled environment, (3) it is inflexible and prone to damage.

Forrest (2004) suggests that a viable alternative may be organic, carbon-based compounds: these materials are obviously cheap, easy-to-manufacture, and can even be flexed, opening up the realm of new uses for photovoltaics. However, the best conversion efficiencies that have been obtained using carbon-based compounds so far have been in the 5% conversion efficiency range (Green, Emery, Hisikawa, and Warta, 2007), which is insufficient for broad market use. (We are not interested in examining the lifetimes of carbon-based photovoltaics in this investigation.)

One hope, however, is that there is no theoretical reason why carbon-based materials cannot be made to have higher conversion efficiencies. So far, the selection of carbon-based candidates has been made mostly through localized domain knowledge and intuition Once a candidate has been chosen, it can either be manufactured and tested to determine its efficiency, or computational simulations can be performed to estimate its theoretical performance as a photovoltaic material. The former is a multi-month-long process, while the later takes on the order of a month to complete. (Adams 2013) These times are obviously too long to make reasonable headway, especially when there are millions of organic materials that we would like to test.

This is where machine learning comes in. If we have a large enough database of molecules that have already had performance tests run over them, we believe that we can develop a learning algorithm that, once trained, can reliably predict the performance of a molecule in the order of, ideally, a few seconds. This problem is particularly interesting because it will hopefully advance simultaneously the way that we use machine learning to learn chemical properties (which has for the most part been domain-specific so far), as well as advance our understanding of applying algorithms in a distributed context to enable faster learning with more systems.

## Adjacency Matrix

The ChemAxon data is enticing, but is also somewhat underwhelming. We chose the features to extract more or less arbitrarily. For many of the features, we have little or no idea what chemical importance they contribute to the HOMO-LUMO gap. Ideally, we would chose features that have direct importance to the quantum-electrical properties of the molecule. We could do further research into "good" features, but that would still be subjective. It would be desirable to model features visible from the molecule itself. The concept of working directly from these observable features and not relying on intermediate models such as those produced by ChemAxon is said to be working from *first principles*. What if we could model features from the atom directly based on its structure?

Let's start by simply visualizing a molecule and thinking about how we can go from there. Let's do the first molecule in our list of SMILES. We can use RDKit to help us extract information directly from the SMILES representation. (iPython has trouble displaying this image directly, so we'll save it and then load it.)

```python
def smiles_image(smiles):
    '''
    Draws a molecule written in SMILES notation. Returns the image.
    '''
    # Represent the molecule cheminformatically
    molecule = Chem.MolFromSmiles(smiles)
    # Draw the annotated molecule
    return Draw.MolToImage(molecule, size=(100, 100), legend=smiles)
```

```python
sample_smile = sample_smiles[0]

showable_smiles_image(sample_smile)
```

As we see, molecules have structure with them. Here, we see that atoms are connected to one another using different types of bonds. "Most" of the bonds in an atom are single bonds. In this molecule, we also observe a double bond between the oxygen and carbon atoms. There's also a carbon ring. This actually represents a series of aromatic bonds, which are alternating single and double bonds which, in total, behave like a 1.5 bond.

With this concept at hand, we could represent the molecule using an adjacency matrix &mdash; atoms sharing a bond would be 'adjacent', and others would not be adjacent. It might also be interesting to model the different bond types. For instance, double bonds could be weighted twice as much as single bonds. We can thus conceptualize two different formulations of the adjacency matrix: a binary adjacency matrix and an $n$ary adjacency matrix. In the binary matrix, matrix elements are determined by bond existence, whereas in the $n$ary adjacency matrix, matrix elements are determined by the number of bonds connecting two atoms. Both of these matrices will clearly be symmetric.

Here's the code to create the matrix.

```python
def smiles_to_adjacency_matrix(smiles_string, binary=False):
  '''
  Converts the SMILES string representation of a molecule into an adjacency
  matrix. The adjacency matrix is a numpy matrix M of values, where M_{i,j}
  represents the adjacency relationship between atoms i and j.

  If binary is False, the M_{i,j} is the number of bonds between atoms i and
  j; e.g. if atom i and j are connected by a single bond, it will be 1, by a
  double bond 2, and no bond 0. If binary is True, then M_{i,j} will be one
  if any type of bond connects atoms and zero otherwise.
  '''
  molecule = Chem.MolFromSmiles(smiles_string)
  num_atoms = molecule.GetNumAtoms()
  adjacency_matrix = np.zeros(shape=(num_atoms, num_atoms))

  for i in range(num_atoms):
    for j in range(num_atoms):
      bond = molecule.GetBondBetweenAtoms(i, j)
      if bond:
        adjacency_matrix[i][j] = 1 if binary else bond.GetBondTypeAsDouble()
      else:
        adjacency_matrix[i][j] = 0

  return np.matrix(adjacency_matrix)
```

## Coulomb Matrix

Adjacency matrices were great for modeling *classical chemistry*. They dealt on the level of bonds and structure. However, HOMO and LUMO are *quantum-chemical concepts* &mdash; that is, we believe that they are determined by total-energy interactions of molecules. This, in some sense, supersedes classical chemistry. We might be interested in modeling the charge interactions of molecules in Cartesian space. Since this is directly related to the quantum charge profile of the molecule, we'd imagine that it may serve as a good basis for HOMO-LUMO prediction.

RDKit makes visualizing charge interactions a snap. Let's consider the charge interactions of our first SMILES.

```python
# Grab the energy contributors. We'll use the atomic number as an indicator for the energies
contribs = [float(molecule.GetAtomWithIdx(i).GetAtomicNum()) for i in range(molecule.GetNumAtoms())]
# Create a matplotlib plot of the energy contours
SimilarityMaps.GetSimilarityMapFromWeights(molecule, contribs, colorMap='Dark2', contourLines=10).show()
```

Modeling quantum-electrical energy interactions may be a promising basis for modeling molecular features. Montavon et al. explore this interesting construct in their 2012 paper. This construct is known as the *Coulomb Matrix* (Montavon et al., 2012). The Coulomb Matrix is an electronic structure representation method based on quantum-mechanical first principles. It only requires the nuclear charges of each atom and their Cartesian coordinates in three dimensional space.
$\newcommand{\bR}{\boldsymbol{R}}$

The Coulomb matrix can be computed simply from the Cartesian coordinates of the atoms and their atomic charges. For atom number $1 \leq i \leq n$ atom, if we let $Z_i$ be its atomic charge and $\bR_i$ be its Cartesian coordinates, then we can specify each $i,j$th element of the $n \times n$ Coulomb matrix as

$$
  C_{i,j} =
    \begin{cases}
      0.5Z_i^{2.4} & \forall i = j \\
      \frac{Z_iZ_j}{|\bR_i - \bR_j|} & \forall i \neq j
    \end{cases}
$$

These attributes can be efficiently computed using Python's RDKit Cheminformatics Tool Suite. Let's look at how we'd do that now.

```python
def smiles_to_coulomb_matrix(smiles_string):
  '''
  Computes the Coulomb matrix representation for a molecule in SMILES
  representation.

  The Coulomb matrix is defined by:

  M_{i,j} = 0.5 * Z_i ^ 2.4   if i = j
          = Z_i * Z_j / (euclidean dist. between molecules i and j)   if i != j

  where Z_i is the nuclear charge for molecule i.

  This Coulomb matrix is returned as a result of this function.
  '''
  molecule = Chem.MolFromSmiles(smiles_string)

  # Prepare the molecule to be optimized using the Universal Force Field
  AllChem.EmbedMolecule(molecule)

  # Use the Universal Force Field to prepare the molecule with a most-likely set
  # of 3D coordinates
  AllChem.UFFOptimizeMolecule(molecule)

  conformer = molecule.GetConformer()
  num_atoms = molecule.GetNumAtoms()

  coulomb_matrix = np.zeros(shape=(num_atoms, num_atoms))

  for i in range(num_atoms):
    for j in range(num_atoms):
      if i == j:
        # Nuclear charge of atom i
        Z_i = molecule.GetAtomWithIdx(i).GetAtomicNum()
        coulomb_matrix[i][j] = 0.5 * Z_i ** 2.4

      else:
        # Nuclear charge of atoms i and j
        Z_i = molecule.GetAtomWithIdx(i).GetAtomicNum()
        Z_j = molecule.GetAtomWithIdx(j).GetAtomicNum()

        # Positions of atoms i and j
        R_i = conformer.GetAtomPosition(i)
        R_j = conformer.GetAtomPosition(j)

        coulomb_matrix[i][j] = \
          Z_i * Z_j / \
          math.sqrt(
            (R_i.x - R_j.x) ** 2 +
            (R_i.y - R_j.y) ** 2 +
            (R_i.z - R_j.z) ** 2
          )

  return np.matrix(coulomb_matrix)
```

## Processing the Matrices

There are a number of issues with the above matrix representations. As mentioned previously, while Neural Networks are good at training over matrices, other kinds of learners are not. This is because we suspect that there are important inter-element dependencies in the matrices. It may be true that if $a$ is adjacent to $b$ and $b$ is adjacent to $c$, then $a$ is more likely to be adjacent to $c$. However, simpler learners cannot automatically extract this feature. One good remedy to this is to summarize the matrix data.

One approach to doing this is to use the *eigenspectrum*. The eigenspectrum is the list of eigenvalues represented in the matrix. These values of course lose some properties of the matrices, but may be good, sparse bases for machine learning. Let's extract the eigenspectra from our matrices above.

```python
def get_eigenvalues(matrix):
    '''
    Computes and returns the eigenvalues in no decreasing order for the provided 
    matrix.
    '''
    return sorted(np.linalg.eigvals(matrix), reverse=True)
```

```python
binary_eigenspectra = [get_eigenvalues(matrix) for matrix in binary_matrices]
nary_eigenspectra = [get_eigenvalues(matrix) for matrix in nary_matrices]
coulomb_eigenspectra = [get_eigenvalues(matrix) for matrix in coulomb_matrices]
```

We'll print out the first few for good measure.

```python
display_list('Binary Eigenspectra', binary_eigenspectra[:3], apply=partial(map, prettyfloat))
display_list('n-ary Eigenspectra', nary_eigenspectra[:3], apply=partial(map, prettyfloat))
display_list('Coulomb Eigenspectra', coulomb_eigenspectra[:3], apply=partial(map, prettyfloat))
```

If you're really forward-thinking about machine learning, something here is probably starting to bother you. These features are all of disparate length! It's very difficult to robustly train learning algorithms with different sized inputs. Thankfully, Montavon et al. (2012) has a solution: when representing a collection of molecules of different number of atoms for machine learning, one should fantasize that they are all comprised of the same number of atoms by padding the molecule with 'null' atoms. These 'null' atoms have zero charge and no bonds. This allows all computed matrices to have the same length.

Let's go ahead and recompute these real quick. We'll be sure to save this data to a pickle file so that we don't have to recompute it every time we restart this notebook.

```python
def pad_matrix(matrix, max_length):
    '''
    For the specified square matrix, this pads the matrix with zero rows
    and columns until it has square length of max_length.
    '''
    pad_length = max_length - matrix.shape[0]
    return np.pad(matrix, ((0, pad_length), (0, pad_length)), mode='constant', constant_values=(0,))
```

```python
# First pass: find the largest matrix
largest = 0
for matrix in binary_matrices:
    largest = max(largest, len(matrix))

print 'The largest matrix has side length', largest

# Pad the matrices
padded_binary_matrices = [pad_matrix(matrix, largest) for matrix in binary_matrices]
padded_nary_matrices = [pad_matrix(matrix, largest) for matrix in nary_matrices]
padded_coulomb_matrices = [pad_matrix(matrix, largest) for matrix in coulomb_matrices]

# Save the matrices
pickle.dump(padded_binary_matrices, open('binary_matrices.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(padded_nary_matrices, open('nary_matrices.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(padded_coulomb_matrices, open('coulomb_matrices.pkl', 'wb'), protocol=pickle.HIGHEST_PROTOCOL)
```

The highly attentive reader might feel uneasy by now. The matrices used above were dependent on the ordering of the atoms. Just because we *labeled* a certain atom as the 'first' in the element, that doesn't mean anything deep. Molecules do not necessarily uniquely identify Coulomb matrices; that is, for an individual molecule, there may be multiple possible Coulomb matrix representations. This is bad for learning, since it means that a regression algorithm may correctly classify two data points differently even if they represent the same molecule.

Three workarounds are proposed for this. These involve:

1. Reducing Coulomb matrices to eigenvalues
2. Applying a deterministic sorting method to Coulomb matrices
3. Randomly permuting Coulomb matrices

Obviously we've already done the first, but if we want to learn from the Coulomb matrices, we'd best also do #2 or #3. We chose to do #2. We sort the matrix row-wise, where rows with a higher "distance" appear earlier. The "distance" is computed by considering the row as a vector and computing its distance.

```python
def sort_matrix(matrix):
    '''
    Sorts and returns a matrix row-wise, where rows with a higher "distance" 
    appear earlier. The "distance" is computed by considering the row as a 
    vector and computing its distance.
    '''
    sorted_rows = sorted(
        [(i, np.linalg.norm(row)) for i, row in enumerate(matrix)], 
        key=itemgetter(1),
        reverse=True
    )
    return np.matrix([matrix[row_number] for i, (row_number, _) in enumerate(sorted_rows)])
```

Now we'll sort all of the matrices.

```python
# Sort all matrices
sorted_binary_matrices = [sort_matrix(matrix) for matrix in binary_matrices]
sorted_nary_matrices = [sort_matrix(matrix) for matrix in nary_matrices]
sorted_coulomb_matrices = [sort_matrix(matrix) for matrix in coulomb_matrices]
```

```python
# 2D PCA projection and visualization
def PCA2vis(x, cluster, title):
    # normalize data
    means = x.mean(axis=0)
    stds = x.std(axis=0)
    x = (x - means)/stds
    
    # PCA
    pca = decomposition.PCA(n_components = 2)
    pca.fit(x)
    print "Explained variance:", pca.explained_variance_ratio_
    P = np.matrix(pca.components_)
    
    # project data points and normalized kmeans centers
    proj = (P * np.matrix(x).T).T
    normalized_centers = (cluster.cluster_centers_ - means)/stds
    projcenters = (P * normalized_centers.T).T
    
    # visualize in 2 dimensions
    colors = np.array(['red','orange','green', 'blue','grey'])
    a = plt.scatter(proj[:,0], proj[:,1], color=dark2_colors, alpha=.6, marker='x')
    b = plt.scatter(projcenters[:,0], projcenters[:,1], color = dark2_colors, marker='o', alpha=0.6)
    c = plt.scatter(projcenters[:,0], projcenters[:,1], color = dark2_colors, marker='o', s=2500, alpha=0.8)
    plt.title(title)
    plt.xlabel('1st principle component projection')
    plt.ylabel('2nd principle component projection')
    plt.legend([a,b],['data projections', 'cluster centers'], frameon=False)
    remove_border()
    
PCA2vis(x, mbk, 'Clustering and PCA on ChemAxon Features')
plt.figure()
PCA2vis(xeig, mbkeig, 'Clustering and PCA on Coulomb Eigenspectra')
```

We can do this for all of our input data types. Note that it's somewhat meaningless to use the matrices for linear regression, so we'll just use the eigenspectra, which may be a good summarization of the matrices in the first place.

```python
def linear_regression_process_data(name_to_learning_methods, name_to_inputs, bold_y):
    '''
    Processes input data of many different formats.
    
    name_to_learning_methods: A dict that carries the name of a learning function to
        a function that creates a sklearn.linear_model.base.LinearRegression instance 
        from input and output data.
    name_to_inputs: A dict that carries the name of an input dataset to the data.
    bold_y: The output for the linear regression learner to target.
    '''
    for learning_name, learning_method in name_to_learning_methods.iteritems():
        for data_name, bold_X in name_to_inputs.iteritems():
            learner = learning_method(bold_X, bold_y)
            coefficients = learner.coef_
            score = learner.score(bold_X, bold_y)
            
            plt.title(data_name + ' Coefficients from\n' + learning_name + \
                      '\nR^2: ' + str(score))
            plt.xlabel('Feature Number')
            plt.ylabel('Coefficient Weight')
            plt.bar(np.arange(len(coefficients)), coefficients, color=dark2_colors[0])
            remove_border()
            
            plt.show()
```

As we've said, the three different approaches produce different $R^2$ values, but that doesn't necessarily mean that one approach is better than the others, since the regularization penalizes how well we can fit in the first place. The correct way to measure these values in a coherent way is to use cross-validation.

Below we use 5-sample cross-validation to perform model comparison. Obviously, we'll only do this for the ChemAxon features representation, as this is the only representation that had reasonable performance for each of the approaches.

First, we'll define a general Mean-Squared Error function for computing the mean squared error of a learner on test data.

```python
def MSE(learner, test_bold_X, test_bold_y):
    '''
    Computes the Mean Squared Error using the trained learner on the 
    provided test data.
    '''
    se = 0.0
    for i in range(len(test_bold_X)):
        se += (learner.predict(test_bold_X[i]) - test_bold_y[i]) ** 2
    return se / len(test_bold_X)
```

Now we want to write a driver for our cross-validations. We want to write this in a way that we can reuse in the future.

```python
def cross_validation_driver(
        bold_X,
        bold_y,
        name_to_data, 
        method_to_create_learner_from_data, 
        num_samples=5):
    '''
    Runs randomized cross validation for a set of input features and output 
    responses.
    
    bold_X: The matrix of input data
    boly_y: The corresponding matrix of output data
    name_to_data: A dict from string keys to arbitrary data. The string keys are
        the names to be used as keys in the returned dict. The values can be anything
        that will be used by the method_to_create_learner_from_data in order to 
        produce a learner for that set of data.
    method_to_create_learner_from_data: This is a function that creates a learner from
        arbitrary data and training data. This function is passed one of the values
        specified in name_to_data and is passed the training input data as a matrix 
        and the training output data as a matrix, and should return a class that 
        supports the predict(X) function as specified in sklean.
    num_samples: How many times we should randomly sample training and test data
    
    Returns a dict. This dict is keyed by the keys of name_to_data. Its values are
        a list of the MSE produced for each learner for each cross validation cycle.
    '''
    # Results will build up our results.
    results = defaultdict(list)

    # Run num_samples iterations of cross-validation.
    for iteration in range(num_samples):
        # Get the cross-validation data
        bold_X_train, bold_X_test, bold_y_train, bold_y_test = \
            cross_validation.train_test_split(
                bold_X, 
                bold_y,
                test_size=1.0 / num_samples
            )
        
        # Compute the MSE for each of the learners
        for name, data in name_to_data.iteritems():
            results[name].append(
                MSE(
                    method_to_create_learner_from_data(data, bold_X_train, bold_y_train),
                    bold_X_test, 
                    bold_y_test
                )
            )
    
    return results
```

The above function is quite complex, but is general enough to act as a driver for many forms of cross-validation. Below we'll show how to use it by computing the cross validation MSEs for our different types of learners.

```python
# Here are our learning methods.
name_to_learning_methods = {
    'Unregularized Linear Regression': unregularized_linear_regression,
    '$\ell_1$-Regularized Linear Regression': l1_linear_regression ,
    '$\ell_2$-Regularized Linear Regression': l2_linear_regression,
}

results = cross_validation_driver(
    sample_features,
    homo_lumo_gaps, 
    name_to_learning_methods,
    lambda learning_method, bold_X_train, bold_y_train: learning_method(bold_X_train, bold_y_train)
)
```

Now that we've got results, we'll make a function to plot the results and then carry out the plot.

```python
def boxplot_dict(d, xlabel='', ymin=None, title='', rotation=10):
    '''
    Produces and displays a boxplot from the provided dict. The string keys
    are the labels, the values are lists of real-valued plot values.
    '''
    plt.title(title)
    plt.xlabel(xlabel)
    pretty_boxplot(plt.boxplot(d.values()))
    plt.ylabel('Mean Squared Error')
    if ymin is not None: plt.ylim(ymin=ymin)
    
    locs, labels = plt.xticks(np.arange(len(d)) + 1, d.keys())
    plt.setp(labels, rotation=rotation)
    
    remove_border()
    plt.show()
```

```python
def print_results_and_averages(results, xlabel='', ymin=None, title='', rotation=10):
    '''
    Shows the boxplot_dict of the results. Below, prints (using print) the
    average MSEs for each entry in the results.
    
    Returns the MSEs, keyed by their name.
    '''
    name_to_MSEs = OrderedDict()
    
    boxplot_dict(results, xlabel=xlabel, ymin=ymin, title=title, rotation=rotation)
    for name, vals in results.iteritems():
        name_to_MSEs[name] = sum(vals) / len(vals)
        print name, 'Average MSE:\t', name_to_MSEs[name]
    print
    
    return name_to_MSEs
```

## Gaussian Processes

Gaussian processes are an advanced *nonparametric* regression construct. This means that they can naturally grow to adapt to additional data, but are in some sense more difficult to interpret from a statistical standpoint. Gaussian processes work by interpolating predictions between observations.

Below are a series of images from http://staffwww.dcs.shef.ac.uk/people/N.Lawrence/gp/ that help demonstrate conceptually how Gaussian processes work. Each pair of images shows how the Gaussian process integrates a new data point into its knowledge. The Gaussian process maintains its amount of uncertainty at each point along the real line; this is the shaded gray area. Larger gray areas therefore represent greater degrees of uncertainty. Whenever a new point is integrated into the Gaussian process, a large amount of uncertainty at that point is explained away. As the number of data points increases to infinity, then, the Gaussian process will become increasingly certain about all points along its distribution.

<img src="http://staffwww.dcs.shef.ac.uk/people/N.Lawrence/gp/demInterpolation3.png" width="50%" /><img src="http://staffwww.dcs.shef.ac.uk/people/N.Lawrence/gp/demInterpolation4.png" width="50%" />
<img src="http://staffwww.dcs.shef.ac.uk/people/N.Lawrence/gp/demInterpolation7.png" width="50%" /><img src="http://staffwww.dcs.shef.ac.uk/people/N.Lawrence/gp/demInterpolation8.png" width="50%" />

Gaussian processes work by considering the covariance between all known points, and estimating the function variance at possible interpolated points. How the covariance between two points is modeled is called the *covariance function* and defines the *covariance matrix* of the Gaussian process. There are several common ways of defining this covariance function:
$\newcommand{\bx}{\boldsymbol{x}}$

1. Squared exponential: $k(\bx_p, \bx_q) = \exp(-\frac{1}{2}|\bx_p - \bx_q|^2)$
2. Absolute exponential: $k(\bx_p, \bx_q) = \exp(-\frac{1}{2}|\bx_p - \bx_q|)$
3. Cubic exponential: $k(\bx_p, \bx_q) = \exp(-\frac{1}{2}|\bx_p - \bx_q|^3)$

We can also add a "nugget" regularization factor $\delta$ to the covariance function, as in $k(\bx_p, \bx_q) + \delta_{p,q}$, in order to regularize the Gaussian process (the nugget technically be scaled by the variance in the input data). This is a necessary step for noisy data! Since we believe that we have noisy data, this is necessary.

The exact implementation details of a Gaussian process are beyond the scope of this course. However, (Bartok, Payne, Kondor, and Csanyi, 2006) suggest that Gaussian processes are a very useful feature for approximating molecular features in chemical space, and so they're definitely a concept that we can't miss out on! Further, `sklearn` has a convenient implementation that we can use out-of-the-box for Gaussian processes.

Now, we have a fairly straightforward task ahead of us of using cross-validation to learn the best values for these parameters. This is very similar to the processes that we used above with linear regression. Technically, we should perform cross validation to determine the model type *and* the nugget value simultaneously. However, a good approximation is to simply perform cross-validation to select the model and then to perform cross-validation to select the nugget value, which significantly reduces the amount of data we have to process. (We will not take this shortcut when processing the larger dataset offline at the end of this section.)

Note that Gaussian processes have *cubic computational complexity* when it comes to prediction! This makes it infeasible to use large datasets and especially infeasible to use large dimensions, so we restrict ourselves to only analyzing the extracted ChemAxon features here.

First, let's look at how the different Gaussian process models perform.

```python
def squared_gp(bold_X, bold_y, corr, nugget=2.2204460492503131e-15):
    '''
    Returns a Gaussian process with the specified covariance profile for the 
    specified input and output features.
    
    (nugget value is the default for machine precision)
    '''
    clf = gaussian_process.GaussianProcess(
        corr=corr, 
        thetaL=1e-3, 
        thetaU=1e0,
        nugget=nugget
    )
    clf.fit(bold_X, bold_y)
    return clf
```

```python
# Here are the covariance method names for our learning methods.
name_to_corr = {
    'Squared Exponential': 'squared_exponential',
    'Absolute Exponential': 'absolute_exponential',
    'Cubic': 'cubic',
}

results = cross_validation_driver(
    sample_features,
    homo_lumo_gaps, 
    name_to_corr,
    lambda corr, bold_X_train, bold_y_train: squared_gp(bold_X_train, bold_y_train, corr)
)

print_results_and_averages(
    results,
    title='Cross-Validated MSE for Different Gaussian Processes',
    rotation=0
);
```

From this cross-validation data, it is clear that we want to go forward with the absolute exponential Gaussian process. Now we want to use cross-validation to compute the best nugget value.

```python
# Here are the nugget values for our learning methods.
name_to_nugget = dict((str(10 ** x), 10 ** x) for x in np.arange(-4, 2))

results = cross_validation_driver(
    sample_features,
    homo_lumo_gaps, 
    name_to_nugget,
    lambda nugget, bold_X_train, bold_y_train: \
        squared_gp(bold_X_train, bold_y_train, 'absolute_exponential', nugget=nugget)
)

# We'll sort the dictionary in order of increasing nugget size.
print_results_and_averages(
    OrderedDict(sorted(dict(results).items(), key=lambda t: t[0])),
    xlabel='Nugget Value',
    title='Cross-Validated MSE for Different Nuggets',
    rotation=0
);
```

From the above plots, it seems like the best Gaussian process is an absolute exponential with a nugget value of 0.01. This is only marginally better than the others, so it's probably safe to make the nugget value *smaller*, if desired. However, making the nugget value *larger* would be a bad idea, as the MSE grows very quickly with largest nugget values.

This is highly encouraging! Remember that smaller nugget values indicate Gaussian processes that are less tolerant to noise. The fact that the nugget value seems to be very small means that our data doesn't have much noise, and suggests that learning may be highly effective over this data.

However, the low nugget value may be an artifact of the small data size. As with linear regression, we have computed the Gaussian processes over a large number of data points offline. Unlike how we were able to use 8000 data points with linear regression, we were only able to manage 1000 at a time with Gaussian processes. This is because Gaussian processes occupy much more memory (since they are a nonparametric model), and have cubic prediction complexity. The results over these 1,000 data points are below.

## Neural Networks

Neural networks are another type of nonparametric learning model. As with Gaussian processes, neural networks are very powerful because they are *nonlinear learners* &mdash; that is, with enough training and data, they are able to learn arbitrary trends in the inputs.

Neural networks have biological origins. They were originally created in the image of the way that the brain works. Neural networks have layers of input, hidden computation and output, and are structured as neurons feeding information from one level to the next using synapses in a complete directed graph from layer to layer, as shown below.

<center>
    <img src="http://www.cerebromente.org.br/n14/mente/knn.gif" /><br />
    Courtesy http://www.cerebromente.org.br/n14/mente/knn.gif
</center>

Neural networks work by first receiving input at the input layer. These are the $\bX$ values from our data. This data is then *fed forward* through the nodes of the network along the weighted synapses. Information is fed forward according to a feed forward function, such as $\tanh(x)$. Once information arrives at the output layer, we compare it to the *actual* value that we are targeting. We take the difference and *back propagate* it using the derivative of our feed forward function, in this case $\frac{\partial}{\partial x}\tanh{x}$. These produce *deltas* at each node, which are the weighted differences of that node from what it "should have been" in order to produce the correct response value. Once these values are propagated all the way back to the input layer, we have a complete set of *weight updates* for each node. We can apply the weight updates along the edges that lead to these nodes, scaled down by some appropriately small *learning rate*, in order to get "closer" to predicting the right value.This whole process is initiated by randomly initializing the synapse edges with weights.

Neural networks are a complex machine learning concept, and for a more rigorous introduction to them, please see Professor Ryan Adam's lecture notes on them on the <a href="http://www.seas.harvard.edu/courses/cs181/S13/">CS181 Course Webpage</a>.

We visualize a single iteration of neural network training below.

<center>
    <img src="http://iamwilchung.files.wordpress.com/2007/03/animate_ann.gif" /><br />
    Courtesy http://iamwilchung.files.wordpress.com/2007/03/animate_ann.gif
</center>

We can repeat this process an arbitrary amount of times, called *epochs*, to get weights that are increasingly better predictors for our response values. With enough training data and time, these predictors can get arbitrarily close to predicting the correct responses!

<center>
    <img src="http://www.codeproject.com/KB/dotnet/predictor/learn.gif" /><br />
    Courtesy http://www.codeproject.com/KB/dotnet/predictor/learn.gif
</center>

Unfortunately, we couldn't find any neural networks in Python packages or on the Internet that suited our needs. We needed a network that was flexible and open source so that we could tinker around with it to parallelize it later in this study. So instead, we rolled our own. Note that since we were able to produce our own matrix, we were able to code it up with a momentum term, which simply means that a fraction of the weights from the last iteration carry through to the next iteration.

See NeuralNetwork.py for our implementation (the code is very long, so we just import it here).

```python
from NeuralNetwork import NeuralNetwork

# An excerpt from the docstring to give some idea of how to use this structure
'''
Serial neural network using back-propagation with momentum. Uses an arbitrary
number of completely-connected hidden layers of arbitrary numbers of nodes
each. Uses an additional input node to represent bias.

Design Notes
------------
After initialization, all layers are represented in the same way.
For instance, all activation values are represented in one list of lists,
  which includes the input, hidden, and output nodes.
All weights are stored in a single list of numpy matrices, where each matrix
  represents the weights from one hidden layer (or the input layer) to the
  next hidden layer (or output layer).
We construct the NeuralNetwork so that it has an additional input node at the
  'bottom' (last input node), which represents the bias term.


CONSTRUCTOR
===========
Initializes and trains the neural network with the specified training
patterns and outputs, using the network structure implied by the hidden
layers.

def __init__(
      self,
      training_patterns,
      outputs,
      hidden_layers,
      weight_initialization_function=lambda layer, num_nodes, num_nodes_next:
        np.matrix(np.random.uniform(-0.2, 0.2,(num_nodes, num_nodes_next))),
      iterations=1000,
      learning_rate=0.2,
      momentum_factor=0.15,
      print_period=100,
      train_immediately=True,
      on_interrupt_save_filename=None):

Parameters
----------
training_patterns: numpy matrix. Each row represents an input pattern.
  IMPORTANT NOTE: The input patterns must be normalized; that is, they must
  represent values between 0 and 1 in order for training to work correctly.
outputs: numpy matrix. The ith row represents the outputs for the ith
  training pattern. IMPORTANT NOTE: The outputs must be normalized; that is,
  they must represent values between 0 and 1 in order for training to work
  correctly.
hidden_layers: List of integers. The length of the list is the number of
  hidden layers. The ith integer in the list is the number of hidden nodes
  in the ith hidden network layer.
weight_initialization_function: A function from the current layer number,
  the number of nodes in this layer, and the number of nodes in the next
  layer to a numpy matrix, whose dimensions are the number of nodes in this
  layer by the number of nodes in the next layer. This function determines
  the weight initializations for the edges from each layer to each next
  layer.
iterations: The integer number of iterations to go through all of the data.
  One iteration consists of feeding forward all of the data and back
  propagating the error rates.
learning_rate: The coefficient in [0, 1] that is the rate to learn from
  back propagation errors.
momentum_factor: The coefficient in [0, 1] that we want to retain the old
  weights when computing the new weights in back propagation.
print_period: The int value that determines how many iterations to wait
  before printing updates to stdout, or None to disable.
train_immediately: True if the NeuralNetwork should train itself using the
  provided data as a part of initialization.
on_interrupt_save_filename: The pickle file to write this NN to when ctrl+C
  is pressed. The network will finish its current training iteration before
  saving.  
''';
```

It's important to note that the inputs and outputs must be normalized to be between zero and one. This is so that the neural network can work with coherent weight values for all features. Before we can use our neural network, let's write a function that normalizes a matrix of inputs columnwise.

When normalizing, we have to be very sure that we can un-normalize later! We have to normalize for training; however, when we run testing, we have to *scale the test values in the same way that we scaled the training values*. This is very important! Also, if we use the neural network for prediction, then we have to un-normalize the resultant predicted value *in the same way that we would un-normalize the response values* in order to have it be in the correct terms.

```python
def normalize_matrix(matrix):
    '''
    Given a numpy matrix of floats, this computes the matrix where each of the
    values is normalized to be between 0.0 and 1.0.
    
    This returns a tuple of the normalized matrix, and the scales and shifts 
    row matrices. In order to "un-normalize" the matrix, you first multiply
    a matrix by scales and then add shifts to it.
    
    Example:
    >>> test = np.matrix([[9,4,-12], [1,8,4], [-4,6,8],[1,2,1]])
    >>> normed, scales, shifts = normalize_matrix(test)
    >>> print normed
    ... [[ 1.      0.3333  0.    ]
         [ 0.3846  1.      0.8   ]
         [ 0.      0.6667  1.    ]
         [ 0.3846  0.      0.65  ]]
    >>> print np.multiply(normed, scales) + shifts
    ... [[  9.   4. -12.]
         [  1.   8.   4.]
         [ -4.   6.   8.]
         [  1.   2.   1.]]
    '''
    # Enforce that it's a matrix of floats of at least two rows
    float_matrix = np.matrix(matrix, dtype=float)
    assert len(float_matrix) > 1, 'Matrix must have at least two rows'
    
    # Shift by subtracting the smallest value from each column
    shifts = float_matrix.min(axis=0)
    shifted = float_matrix - shifts
    
    # Scale by dividing by the largest value in each column
    scales = shifted.max(axis=0)
    normed = np.nan_to_num(shifted / scales)
    
    return normed, scales, shifts
```

```python
features_normed, features_scales, features_shifts = normalize_matrix(sample_features)
gaps_normed, gaps_scales, gaps_shifts = normalize_matrix(np.matrix(homo_lumo_gaps).T)
```

Throughout the below discussions, we use the variable $m$ to represent the total number of processors. $p$ is the number of data dimensions. $1 \leq k \leq m$ refers to an arbitrary processor. Variables subscripted with a $(k)$ refer to machine $k$'s local copy of that variable.

## Distributed Linear Regression

Distributed linear regression can be divided into four phases:

<center>
    <img src="files/Distributed Linear Regression.png" /><br />
</center>
$\newcommand{\bbeta}{\boldsymbol{\beta}}$
$\newcommand{\bone}{\boldsymbol{1}}$
$\newcommand{\co}          {\overline}$
$\newcommand{\trans}       {\text{T}}$
$\newcommand{\IR}{\text{InResp}}$
$\newcommand{\OR}{\text{OutResp}}$

### Phase 1

This is the data distribution phase. The root node reads the $n$ data points as the $n \times p$ matrix of regressors $\bX$ and the $n \times 1$ vector of response variables $\by$. The root splits the data into $m$ approximately-equally-sized matrices and distributes it to the other nodes using MPI's **Scatter**.


### Phase 2

This phase completes all computation that can be done without sharing information. The information computed in this phase summarizes the data for succinct computation by other processors in the next phase. Node $k$ with $n_k$ data points computes:

* The vector of regressor column means, $\co\bx_{(k)} = \bX_{(k)}^\trans\bone/n_{(k)}$
* The centered $\bX_{(k)}$ matrix; that is, the matrix of regressor differences from means, $\co\bX_{(k)}=\bX_{(k)}-\bone\co\bx_{(k)}$
* The output mean, $\co{y}_{(k)} = \by_{(k)}^\trans\bone/n_{(k)}$
* The vector of output differences from the output mean, $\co\by_{(k)} = \by_{(k)} - \bone\co{y}_{(k)}$
* The squared centered regressors, $\co\bX_{(k)}^\trans\co\bX_{(k)}$
* The total centered regressors-times-centered outputs, $\co\bX_{(k)}^\trans\co\by_{(k)}$
* This processor's responsibility for the regressors, $\IR_{(k)} = n_k\co\bx_{(k)}/n$
* This processor's responsibility for the response variables, $\OR_{(k)} = n_k\co y_{(k)}/n$


### Phase 3

Using MPI's **AllReduce**, we efficiently sum up the input and output responsibilities across all processors onto all processors. The sum of the $\IR_k$'s is clearly $\co\bx$, and the sum of the $\OR_k$'s is $\co y$. Each processor now computes $n_k (\co\bx_{(k)} - \co\bx) (\co y_{(k)} - \co y)$ and $n_k (\co\bx_{(k)} - \co\bx) (\co\bx_{(k)} - \co\bx)^\trans$. Realize that each processor can now compute $\co\bX_{(k)}^\trans\co\by_{(k)} + n_k (\co\bx_{(k)} - \co\bx) (\co y_{(k)} - \co y)$, the processor's fraction of $\co\bX^\trans\co\by$, and $\co\bX_{(k)}^\trans\co\bX_{(k)} + n_k (\co\bx_{(k)} - \co\bx) (\co\bx_{(k)} - \co\bx)^\trans$, the processor's fraction of $\co\bX^\trans\co\bX$.


### Phase 4

Using MPI's **Reduce** on the root to sum up remote values across processors, we can easily compute $\co\bX^\trans\co\by$ by summing over all $\co\bX_{(k)}^\trans\co\by_{(k)} + n_k (\co\bx_{(k)} - \co\bx) (\co y_{(k)} - \co y)$ and $\co\bX^\trans\co\bX$  by summing over all $\co\bX_{(k)}^\trans\co\bX_{(k)} + n_k (\co\bx_{(k)} - \co\bx) (\co\bx_{(k)} - \co\bx)^\trans$.


### Final Compute

 Finally, we can compute the maximum likelihood weights for linear regression. For linear regression of the form $y = \beta_0 + \bbeta_1 \bx$, we have that
 
$$\hat\bbeta_1 = (\co\bX^\trans\co\bX)^{-1}\co\bX^\trans\co\by$$
$$\hat\beta_0 = \co y-\co\bx^\trans\hat\bbeta_1$$


### Notes

Note for the final compute that $\co\bX^\trans\co\bX$ is a square matrix that is the size of the number of feature dimensions squared (thus independent of $n$), so is easy to invert. It is also easy to multiply this by $\co\bX^\trans\co\by$, which is a column vector the size of the number of feature dimensions. We can add arbitrary regularization of magnitude $\lambda$ to the linear regression by adding $\lambda I$ to $\co\bX^\trans\co\bX$ before inversion, although for this segment of the project we used unregularized linear regression, since it proved most promising on the small-scale samples.

## Distributed Neural Networks

Distributed neural networks can be divided into three phases:

<center>
    <img src="files/Distributed Neural Networks.png" /><br />
</center>

### Phase 1
In this algorithm, the entire problem must be replicated on *all* nodes. We have each node read in the entire dataset, and use a common random seed across the network (just for initialization) to replicate exactly the network on each node.

### Phase 2
Each processor selects a *random* subset of $1.5 / m$ of the data (different amounts can be used). Each neural network computes, but does not apply, the batch weight updates.

### Phase 3
Using MPI's **AllReduce**, sums of all batch weight updates are sent to all processors in the network. Each processor then updates its recorded original weights from before the last training iteration by adding these batch weight updates. We then repeat from Phase 2.

### Final compute
The above process is computed as long as desired. Note that the state of each neural network is identical after every weight exchange, so we can simply use the neural network on one of the machines at the end of the process as the output of the algorithm. We can also estimate the global error rate by summing the local error rates across all machines when exchanging weights.

### Implementation

We wrote a driver for this distributed neural network problem, which can be run on Resonance. The code is listed here, although will only be runnable on Resonance.

```python
# Again, see parallel.py for our implementation of parallelized neural networks.
from parallel import run_parallel_neural_network

'''
Runs regression using a neural network. The specific algorithm is inspired by
the Pattern Parallel Training algorithm, outlined by a 2008 paper by Dahl,
McAvinney, and Newhall. This must be run on an MPI-enabled distributed
cluster. This returns the resultant network on the root processor.

CONSTRUCTOR
===========
def run_parallel_neural_network(filename):


PHASE 1: PRE-DISTRIBUTED PROCESSING

The algorithm requires that all data be present on all machines. Therefore,
rather than have the root process read and distribute the data, we have the
all processes read the data independently. Since these are reads, they can be
done in parallel without any fear of data mangling or inconsistency.


PHASE 2: DISTRIBUTE DATA AND PERFORM PARALLEL COMPUTATIONS

Here we want to run the distributed version of the neural network. This is
done in a few steps, so we'll start slowly:

We iterate for a number of epochs. This can be determined by the mean
squared error, but in practice, since the error rate decays over time, we run
for 'as long as possible', so it's reasonable to specify this for a number of
epochs. Each iteration, we choose a random subset of the data to process.

For each iteration, we want to record the weight updates for that iteration.
This is so that we can broadcast these weight updates to the other processors.
The serial neural network algorithm that we built doesn't quite do what we
want, since it's optimized to run for many training trials. However, since
we've exposed some of the otherwise internal neural network API, we can take
over here to do what we want. Specifically, we want to record the weights
before each training session, train, and then record the weights after. By
subtracting the weights after training from the weights before training, we
can get all of the weight update matrices for each layer. We can then
broadcast and sum these matrices to all processors using MPI's AllReduce.
These weight updates can be applied to each processes *PRE-TRAINING* weights
to achieve a *globally-consistent* set of post-training weight matrices. It's
also simple to record the new momentum matrices, since they literally are the
update matrices that each processor is receiving.

By repeating the above process, we can run multiple distributed training
epochs of the NeuralNetwork. It's important that all of the distributed
weight matrices remain globally consistent, so every few transmission
iterations, we check to ensure that all of the weight matrices are, within a
very small margin of error, the same as the root process's weight matrix.


PHASE 3: COLLECT DATA & PERFORM SERIAL FOLLOW-UP COMPUTATIONS

At this stage, we have a globally-consistent set of weight matrices that have
arisen from performing a distributed training of the NeuralNetwork. Since
these weight matrices are globally consistent, we actually don't have any
follow-up collection of distributed data to perform, and can use the computed
weight matrices as a basis for the neural network.
''';
```

This code uses a CSV file in which all columns except the last are input features. The last column is interpreted as the target value.

This has been a long adventure! We've walked through many disparate feature representations, a variety of data explorations both useful and less so, and several different learning approaches on the small scale and the massive. Overall, the results are very encouraging. Not only have we identified effective approaches to modeling the HOMO-LUMO gap from a basic molecular representation, but we've also identified the key promising areas that deserve additional research. We'll recap each of the major themes of this project and suggest take-aways.


#### Representation

We looked at extracted features, adjacency matrices, coulomb matrices, and eigenspectra. The adjacency representations consistently had pathological failures or were otherwise bad, and so we dropped them early on as a viable model. The eigenspectra were really only good as a summarization for raw matrices. We would have preferred to learn directly over the full matrix representations, but without a good way of extracting inter-element dependencies from this matrix, it's not viable for most forms of regression.

Coulomb matrices and extracted features, on the other hand, seemed to be very promising! Extracted features consistently performed well across-the-board. This is probably because these extracted features used intermediate chemical computations to partially summarize the relevant information inherent in the molecule. Even if we don't know what the features mean, that doesn't mean that they don't make for great regressors! An interesting line of additional research would be to explore more and alternate features, or to perform variable selection to determine the most "valuable" features. Coulomb matrices performed *very* well with neural networks. This is probably because Coulomb matrices are modeling the first principles interactions in the molecule, and therefore may contain most or all of the information that we need to learn HOMO-LUMO values.


#### Exploration

Data exploration proved to be challenging due to the nature of our representations. Cheminformatic properties and structure based matrices yield complex, high dimensional features that are difficult to visualize and reason about. Clustering and PCA projections were not particularly helpful, but just visualizing the distributions of each dimension at least helps give a feel for the scales of what we are working with. This told us that normalization of features would be important. In addition, looking at the correlation scatter matrix lets us see potential interactions between different features, which could help us understand relationships between important properties that our models might discover.


#### Regression

We examined Gaussian processes in serial, and linear regression and neural networks in serial and parallel. Gaussian processes seemed to perform the best of the serial models, but prove very difficult to parallelize. This limitation makes it difficult to say how well they will generalize to the entire dataset. Nonetheless, since they are a nonparametric model, we suspect that they would perform well with data scaling.

Linear regression is interesting. It is fast with low memory requirements, and therefore is highly amenable to scaling. On the small scale, it performed reasonably well to warrant further testing, but after parallelizing, the results appeared significantly worse than the serial version. This is suggestive that linear regression is a poor model for the actual underlying regression.

Although our final results for neural networks were not the best of all of the learners used, they were nontheless highly encouraging. Neural networks not only scaled well, but it seems that we may be able to get significantly better performance results just by training for additional time. This is certainly a direction for additional research. Furthermore, they allow us to effectively use the Coulomb matrix directly as a feature representation. This seems to perform remarkably well, with potential to perform even better with additional scaling.
